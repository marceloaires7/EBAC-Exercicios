# Métodos hierárquicos de agrupamento - Tarefa

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

import scipy.cluster.hierarchy as shc

%reload_ext watermark
%watermark -iv -v -p sklearn
```

## 1) Repetindo o que fizemos em aula

Primeiramente vamos repetir o que fizemos em aula com a base de pinguins.

1. Selecione as variáveis quantitativas da base de pinguins
2. Elimine valores faltantes
3. Padronize as variáveis para média zero e desvio padrão 1
4. Faça um agrupamento hierárquico dividindo essa base em 3 grupos
5. Faça o Dendrograma

```python
# Lendo dataset 'penguins' do seaborn.
peng = sns.load_dataset('penguins')

# Definindo nome do index como 'id'.
peng.index.name='id'

# Extraindo somente variáveis quantitativas dataset e retirando as variaveis vazias (na).
peng_num = peng.select_dtypes('number').dropna()

# Padronizando as variáveis numéricas.
peng_pad = pd.DataFrame(StandardScaler().fit_transform(peng_num), columns=peng_num.columns)

# Agrupamento hierárquico dividindo essa base em 3 grupos.
clus = AgglomerativeClustering(linkage            = 'complete',
                               distance_threshold = None,
                               n_clusters         = 3).fit(peng_pad)

# Amostra 'peng_pad'
peng_pad
```

```python
# Acrescentando a variável 'grupo_3' no dataframe 'peng_num'.
peng_num['grupo_3'] = clus.labels_
peng_num
```

```python
# Acrescentando a variável 'grupo_3' no dataframe 'peng'.
peng = peng.merge(peng_num['grupo_3'], how='left', on='id')
peng
```

```python
# Pairplot do dataset 'peng' separados por grupo_3.
sns.pairplot(data=peng, hue='grupo_3', palette="tab10")
```

```python
# Dendrograma do dataset 'peng_pad' mostrando a divisão em 3 grupos.
plt.figure(figsize=(15, 15))

plt.title("Customer Dendograms")
shc.set_link_color_palette(['#1F77B4','#2CA02C', '#FF7F0E'])
dend_3 = shc.dendrogram(shc.linkage(peng_pad, method='complete'), above_threshold_color='grey')
```

## 3) Alterando o agrupamento

O Dendrogarma sugere que podemem haver grupos interessantes se solicitarmos mais que 3 grupos. Se quisermos testar o agrupamento com mais de 3 grupos, quantos grupos o Dendrograma te sugere? (Dica: não faça mais que 6 grupos).

    - Quantos pinguins de cada espécie machos e fêmeas há em cada grupo?

```python
# Agrupamento hierárquico dividindo essa base em 6 grupos.
clus_6 = AgglomerativeClustering(linkage            = 'complete',
                               distance_threshold = None,
                               n_clusters         = 6).fit(peng_pad)
```

```python
# Acrescentando a variável 'grupo_6' no dataframe 'peng_num'.
peng_num['grupo_6'] = clus_6.labels_
peng_num
```

```python
# Acrescentando a variável 'grupo_6' no dataframe 'peng'.
peng = peng.merge(peng_num['grupo_6'], how='left', on='id')
peng
```

```python
# Pairplot do dataset 'peng' separados por grupo_6.
sns.pairplot(data=peng.drop(columns='grupo_3'), hue='grupo_6', palette="tab10")
```

### Analise esses agrupamentos:<br>- Quantos pinguins de cada espécie há em cada grupo?

```python
# Crosstab entre 'species' e 'grupo_6'.
pd.crosstab(peng['species'],
            peng['grupo_6'],
            margins=True,
            margins_name= 'Total')
```

### Analise esses agrupamentos:<br>- Quantos pinguins de cada espécie machos e fêmeas há em cada grupo?

```python
peng[['species', 'sex']].value_counts()
```

```python
# Análise de 'species' e 'sex' em relação ao 'grupo_6'
species_sex = peng[['species', 'sex', 'grupo_6']].value_counts().unstack()
species_sex.columns = ['Gentoo_Male', 'Adelie_Female', 'Adelie_Male', 'Gentoo_Female', 'Chinstrap_Female', 'Chinstrap_Male']
species_sex.sort_index(axis=1).style.highlight_null(color='black').highlight_quantile(color='orange').highlight_max(color='green', axis = 1).format(precision=2, decimal=',')
```

```python
# Dendrograma do dataset 'peng_pad' mostrando a divisão em 6 grupos.
plt.figure(figsize=(15, 15))

plt.title("Customer Dendograms")
shc.set_link_color_palette(['#D62728','#1F77B4', '#8C564B', '#9467BD', '#FF7F0E', '#2CA02C'])
dend_6 = shc.dendrogram(shc.linkage(peng_pad, method='complete'), above_threshold_color='gray', color_threshold=3.4)
```

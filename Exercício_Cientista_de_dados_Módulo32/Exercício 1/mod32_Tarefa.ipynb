1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
# Mdia
mu = 15

# Desvio Padro
sigma = 0.1

# Intervalo low (15,2 kg) e high (infinito)
low = 15.2
high = np.inf

# Calcular probabilidade entre low e high
prob_area = stats.norm.cdf(high, mu, sigma) - stats.norm.cdf(low, mu, sigma)
print(f'A probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg  de {prob_area:.2%}'.replace('.',','))
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Peso em que o percentil  de ordem 95%.

# Percentil
perc = 0.95

# Peso 
prob_peso = stats.norm.ppf(perc, mu, sigma)
print(f'A probabilidade de um pacote ter peso menor que {prob_peso:.2f} kg  de 95%')
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu cdigo comea aqui

# Mdia
mu = 0

# Grau de Liberdade
gl = 5

# Varincia e Desvio padro
variance = 1
sigma = math.sqrt(variance)

# x e y
x = Y
y = stats.norm.pdf(x, mu, sigma)

# Intervalo low (-6) e high (0)
low = -6
high = 0

# Probabilidade Normal e t-studdent
prob_normal = stats.norm.cdf(x, mu, sigma)
prob_t = stats.t.cdf(x, gl)

print(f'Normal Padro:\n\n {prob_normal[:10]}\n\n########################################################################\n')
print(f'5 Graus de Liberdade:\n\n {prob_t[:10]}')
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
# Definindo subplots dos grficos.
fig, ax = plt.subplots(nrows=2, figsize=(10,10))

# Plot Grfico 1.
ax[0].plot(x, y, label='Normal')
t = stats.t.pdf(x, gl)
ax[0].plot(x, t, label=f'gl = {gl}')

# Plot Grfico 2.
ax[1].plot(x, prob_normal, label='Normal')
ax[1].plot(x, prob_t, label=f'gl = {gl}')

# Legendas.
ax[0].legend()
ax[1].legend()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Seu cdigo comea aqui

# Mdia
mu = 0

# Grau de Liberdade
gl = 5

# Varincia e Desvio padro
variance = 1
sigma = math.sqrt(variance)

# Percentis correspondentes aos valores de probabilidade acumulada
perc_normal = stats.norm.ppf(Y, mu, sigma)
perc_t = stats.t.ppf(Y, gl)

print(f'Percentil Normal Padro:\n\n {perc_normal}\n\n########################################################################\n')
print(f'Percentil 5 Graus de Liberdade:\n\n {perc_t}')
```

6. Faa um grfico com os dados do item anterior.

```python
# Definindo subplots dos grficos.
fig, ax = plt.subplots(figsize=(10, 4))

# x
x = Y

# Plot Grfico.
ax.plot(x, perc_normal, label='Normal')
ax.plot(x, perc_t, label=f'gl = {gl}')

# Legendas
ax.legend()
```

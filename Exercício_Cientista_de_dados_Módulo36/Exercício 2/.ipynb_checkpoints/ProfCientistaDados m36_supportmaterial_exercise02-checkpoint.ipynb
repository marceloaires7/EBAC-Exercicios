**OBJETIVO:** Algumas notas no do retorno, o que pode ocorrer por diversos motivos, um deles seria a ocorrncia de produtos no incentivados como cigarros e bebidas alclicas. O nosso objetivo  tentar prever que tipo de nota tem maior ou menor propenso a fornecer crditos.

**Observao**: Esta  uma base real, havendo caractersticas de acordo. Pode haver uma variabilidade difcil de explicar, em parte por causa de uma freqncia de notas inconstante no tempo (por diversos motivos), em parte por haverem informaes importantes, protegidas pela LGPD, no constantes na base, que pode interferir no crdito gerado, o que vai ocasionar uma variabilidade difcil de se explicar. *Welcome to the jungle*, A vida real  assim :)

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme()

%reload_ext watermark
%watermark -iv -v

import warnings

warnings.filterwarnings("ignore")
````

````python
def analise(data, y):
    analise = pd.DataFrame({'dtype': data.dtypes,
                            'contagem': data.count(),
                            'missing': data.isna().sum(),
                            'nunique': data.nunique(),
                            'papel': 'covariavel'})
    analise.loc[analise.index == y, 'papel'] = 'resposta'
    return analise

def IV(variavel, resposta):
    tab = pd.crosstab(variavel, resposta, margins=True, margins_name='total')

    rtulo_nao_evento = tab.columns[0]
    rtulo_evento = tab.columns[1]

    tab['pct_evento'] = tab[rtulo_evento]/tab.loc['total',rtulo_evento]
    tab['pct_nao_evento'] = tab[rtulo_nao_evento]/tab.loc['total',rtulo_nao_evento]
    tab['woe'] = np.log(tab.pct_evento/tab.pct_nao_evento)
    tab['iv_parcial'] = (tab.pct_evento - tab.pct_nao_evento)*tab.woe
    return tab
````

````python
df = pd.read_pickle('./Dados/base_nfp.pkl')
df.head()
````

````python
df['Retorno'].fillna(0, inplace=True)
df = df.loc[df['Retorno'] < 800]
df.reset_index(inplace=True, drop=True)

analise(data=df, y='flag_credito')
````

## Criando variveis

**Variveis temporais**:  possvel que haja alguma regra de liberao de crditos de acordo com o dia do ms, o dia da semana, fim de semana (sim ou no), ou at alteraes temporais da regra. Vamos criar as seguintes variveis para testar estas hipteses:

- Dia da semana
- Dummy para fim de semana
- Dia do ms
- Ms (j na base)
- Trimestre
- Ano (j na base)

**Valor da nota**: Vamos criar tambm algumas variveis para resumir o valor da nota, e posteriormente criar uma intuio sobre categorizaes de variveis contnuas.

- Crie categorizaes da varivel ```Valor NF```, por quants, em 5, 10, 20 e 50 categorias.

````python
# Variveis Temporais
df['Dia_da_Semana'] = df['Data Emisso'].dt.weekday
df['Fim_de_Semana'] = df['Dia_da_Semana'].isin([5,6]).astype('int')
df['Dia_do_mes']    = df['Data Emisso'].dt.day
df['Mes']           = df['Data Emisso'].dt.month
df['Trimestre']     = df['Data Emisso'].dt.to_period('Q')

df[['Dia_da_Semana',
    'Fim_de_Semana',
    'Dia_do_mes',
    'Mes',
    'Trimestre',
    'Ano']].nunique().to_frame('Quantidade')
````

````python
# Valor da Nota

df['Valor_NF_Cat_5'] = pd.qcut(df['Valor NF'], q=5, labels=False)
df['Valor_NF_Cat_10'] = pd.qcut(df['Valor NF'], q=10, labels=False)
df['Valor_NF_Cat_20'] = pd.qcut(df['Valor NF'], q=20, labels=False)
df['Valor_NF_Cat_50'] = pd.qcut(df['Valor NF'], q=50, labels=False)

df[['Valor_NF_Cat_5',
    'Valor_NF_Cat_10',
    'Valor_NF_Cat_20',
    'Valor_NF_Cat_50']].nunique().to_frame('Quantidade')
````

## Calcule o IV de todas essas variveis

Sugesto: Construa um metadados e tente utilizar uma funo como a que foi utilizada durante a aula. Voc vai conseguir fazer esse clculo para as variveis uma a uma, mas em casos reais em que vamos mais alm na criao de variveis, podemos ter centenas, ou at milhares de variveis, esses clculos *precisam* ser automticos nesses casos.

````python
df_tratado = df.drop(columns=['CNPJ emit.', 'No.', 'Emitente', 'Data Emisso', 'Data Registro', 'Situao do Crdito'])
metadados = analise(data=df_tratado,y='flag_credito')
metadados
````

````python
for var in metadados[metadados.papel=='covariavel'].index:
    if  (metadados.loc[var, 'nunique']>50):
        metadados.loc[var, 'IV'] = IV(pd.qcut(df_tratado[var],5,duplicates='drop'), df_tratado.flag_credito)['iv_parcial'].sum()
    else: 
        metadados.loc[var, 'IV'] = IV(df_tratado[var], df_tratado.flag_credito)['iv_parcial'].sum()
    
metadados.sort_values(by='IV', ascending=False, inplace=True)
metadados
````

## Conclua

- Quais variveis parecem apresentar maior poder de discriminao segundo o IV?

**Resposta**: 'categoria': 0.408 (Forte)

- Quando categorizamos variveis contnuas parece haver alguma relao entre o nmero de categorias e o IV?

**Resposta**: Vemos que quanto maior o nmero de categorias, melhor  o IV.

| Valor_NF_Cat_50 | Valor_NF_Cat_20 | Valor_NF_Cat_10 | Valor_NF_Cat_5 | Valor NF |
| --------------- | --------------- | --------------- | -------------- | -------- |
|      0.152      |      0.125      |      0.107      |      0.057     |   0.057  |
|     (Mdio)     |     (Mdio)     |     (Mdio)     |     (Fraco)    |  (Fraco) |

````python
print(f'\033[1;7mInformation Value (IV):\033[0m\n{"-"*36}')

for i, j in zip(metadados.index, metadados.IV):
    if j > 0.5:
        print(f'\033[1m{i}:\033[0m\033[1;91m {j:.3f} (Suspeito de to alto)\033[0m\n{"-"*36}')
    elif j > 0.3:
        print(f'\033[1m{i}:\033[0m\033[1;94m {j:.3f} (Forte)\033[0m\n{"-"*36}')
    elif j > 0.1:
        print(f'\033[1m{i}:\033[0m\033[1;92m {j:.3f} (Mdio)\033[0m\n{"-"*36}')
    elif j > 0.02:
        print(f'\033[1m{i}:\033[0m\033[1;93m {j:.3f} (Fraco)\033[0m\n{"-"*36}')
    else:
        print(f'\033[1m{i}:\033[0m\033[1;91m {j:.3f} (Intil)\033[0m\n{"-"*36}')
````

## Avalie o WOE e o IV no tempo

Avalie o WOE e o IV no tempo da varivel ```categoria```.

````python
WOE_cat_tempo = []
IV_cat_tempo = []

trimestres = df['Trimestre'].sort_values().unique()

for trimestre in trimestres:
    df_trimestre = df[df['Trimestre'] == trimestre]
    tab = IV(df_trimestre['categoria'], df_trimestre['flag_credito'])
    WOE_cat_tempo.append(tab['woe'])
    IV_cat_tempo.append(tab['iv_parcial'])

WOE_cat_tempo = pd.DataFrame(WOE_cat_tempo, index=trimestres).drop('total', axis=1)
IV_cat_tempo = pd.DataFrame(IV_cat_tempo, index=trimestres).drop('total', axis=1).sum(axis=1).to_frame('IV')
WOE_cat_tempo.loc['total', :] = WOE_cat_tempo.sum()
WOE_cat_tempo.sort_values(by='total', ascending=False, axis=1, inplace=True)
WOE_cat_tempo
````

````python
print(f'\033[1;7mInformation Value (IV):\033[0m\n{"-"*36}')

for i, j in zip(IV_cat_tempo.index, IV_cat_tempo.IV):
    if j > 0.5:
        print(f'\033[1m{i}:\033[0m\033[1;91m {j:.3f} (Suspeito de to alto)\033[0m\n{"-"*36}')
    elif j > 0.3:
        print(f'\033[1m{i}:\033[0m\033[1;94m {j:.3f} (Forte)\033[0m\n{"-"*36}')
    elif j > 0.1:
        print(f'\033[1m{i}:\033[0m\033[1;92m {j:.3f} (Mdio)\033[0m\n{"-"*36}')
    elif j > 0.02:
        print(f'\033[1m{i}:\033[0m\033[1;93m {j:.3f} (Fraco)\033[0m\n{"-"*36}')
    else:
        print(f'\033[1m{i}:\033[0m\033[1;91m {j:.3f} (Intil)\033[0m\n{"-"*36}')
````

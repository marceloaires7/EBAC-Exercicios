**OBJETIVO:** Algumas notas não dão retorno, o que pode ocorrer por diversos motivos, um deles seria a ocorrência de produtos não incentivados como cigarros e bebidas alcólicas. O nosso objetivo é tentar prever que tipo de nota tem maior ou menor propensão a fornecer créditos.

**Observação**: Esta é uma base real, havendo características de acordo. Pode haver uma variabilidade difícil de explicar, em parte por causa de uma freqüência de notas inconstante no tempo (por diversos motivos), em parte por haverem informações importantes, protegidas pela LGPD, não constantes na base, que pode interferir no crédito gerado, o que vai ocasionar uma variabilidade difícil de se explicar. *Welcome to the jungle*, A vida real é assim :)

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme()

%reload_ext watermark
%watermark -iv -v

import warnings

warnings.filterwarnings("ignore")
````

````python
def analise(data, y):
    analise = pd.DataFrame({'dtype': data.dtypes,
                            'contagem': data.count(),
                            'missing': data.isna().sum(),
                            'nunique': data.nunique(),
                            'papel': 'covariavel'})
    analise.loc[analise.index == y, 'papel'] = 'resposta'
    return analise

def IV(variavel, resposta):
    tab = pd.crosstab(variavel, resposta, margins=True, margins_name='total')

    rótulo_nao_evento = tab.columns[0]
    rótulo_evento = tab.columns[1]

    tab['pct_evento'] = tab[rótulo_evento]/tab.loc['total',rótulo_evento]
    tab['pct_nao_evento'] = tab[rótulo_nao_evento]/tab.loc['total',rótulo_nao_evento]
    tab['woe'] = np.log(tab.pct_evento/tab.pct_nao_evento)
    tab['iv_parcial'] = (tab.pct_evento - tab.pct_nao_evento)*tab.woe
    return tab
````

````python
df = pd.read_pickle('./Dados/base_nfp.pkl')
df.head()
````

````python
df['Retorno'].fillna(0, inplace=True)
df = df.loc[df['Retorno'] < 800]
df.reset_index(inplace=True, drop=True)

analise(data=df, y='flag_credito')
````

## Criando variáveis

**Variáveis temporais**: É possível que haja alguma regra de liberação de créditos de acordo com o dia do mês, o dia da semana, fim de semana (sim ou não), ou até alterações temporais da regra. Vamos criar as seguintes variáveis para testar estas hipóteses:

- Dia da semana
- Dummy para fim de semana
- Dia do mês
- Mês (já na base)
- Trimestre
- Ano (já na base)

**Valor da nota**: Vamos criar também algumas variáveis para resumir o valor da nota, e posteriormente criar uma intuição sobre categorizações de variáveis contínuas.

- Crie categorizações da variável ```Valor NF```, por quantís, em 5, 10, 20 e 50 categorias.

````python
# Variáveis Temporais
df['Dia_da_Semana'] = df['Data Emissão'].dt.weekday
df['Fim_de_Semana'] = df['Dia_da_Semana'].isin([5,6]).astype('int')
df['Dia_do_mes']    = df['Data Emissão'].dt.day
df['Mes']           = df['Data Emissão'].dt.month
df['Trimestre']     = df['Data Emissão'].dt.to_period('Q')

df[['Dia_da_Semana',
    'Fim_de_Semana',
    'Dia_do_mes',
    'Mes',
    'Trimestre',
    'Ano']].nunique().to_frame('Quantidade')
````

````python
# Valor da Nota

df['Valor_NF_Cat_5'] = pd.qcut(df['Valor NF'], q=5, labels=False)
df['Valor_NF_Cat_10'] = pd.qcut(df['Valor NF'], q=10, labels=False)
df['Valor_NF_Cat_20'] = pd.qcut(df['Valor NF'], q=20, labels=False)
df['Valor_NF_Cat_50'] = pd.qcut(df['Valor NF'], q=50, labels=False)

df[['Valor_NF_Cat_5',
    'Valor_NF_Cat_10',
    'Valor_NF_Cat_20',
    'Valor_NF_Cat_50']].nunique().to_frame('Quantidade')
````

## Calcule o IV de todas essas variáveis

Sugestão: Construa um metadados e tente utilizar uma função como a que foi utilizada durante a aula. Você vai conseguir fazer esse cálculo para as variáveis uma a uma, mas em casos reais em que vamos mais além na criação de variáveis, podemos ter centenas, ou até milhares de variáveis, esses cálculos *precisam* ser automáticos nesses casos.

````python
df_tratado = df.drop(columns=['CNPJ emit.', 'No.', 'Emitente', 'Data Emissão', 'Data Registro', 'Situação do Crédito'])
metadados = analise(data=df_tratado,y='flag_credito')
metadados
````

````python
for var in metadados[metadados.papel=='covariavel'].index:
    if  (metadados.loc[var, 'nunique']>50):
        metadados.loc[var, 'IV'] = IV(pd.qcut(df_tratado[var],5,duplicates='drop'), df_tratado.flag_credito)['iv_parcial'].sum()
    else: 
        metadados.loc[var, 'IV'] = IV(df_tratado[var], df_tratado.flag_credito)['iv_parcial'].sum()
    
metadados.sort_values(by='IV', ascending=False, inplace=True)
metadados
````

## Conclua

- Quais variáveis parecem apresentar maior poder de discriminação segundo o IV?

**Resposta**: 'categoria': 0.408 (Forte)

- Quando categorizamos variáveis contínuas parece haver alguma relação entre o número de categorias e o IV?

**Resposta**: Vemos que quanto maior o número de categorias, melhor é o IV.

| Valor_NF_Cat_50 | Valor_NF_Cat_20 | Valor_NF_Cat_10 | Valor_NF_Cat_5 | Valor NF |
| --------------- | --------------- | --------------- | -------------- | -------- |
|      0.152      |      0.125      |      0.107      |      0.057     |   0.057  |
|     (Médio)     |     (Médio)     |     (Médio)     |     (Fraco)    |  (Fraco) |

````python
print(f'\033[1;7mInformation Value (IV):\033[0m\n{"-"*36}')

for i, j in zip(metadados.index, metadados.IV):
    if j > 0.5:
        print(f'\033[1m{i}:\033[0m\033[1;91m {j:.3f} (Suspeito de tão alto)\033[0m\n{"-"*36}')
    elif j > 0.3:
        print(f'\033[1m{i}:\033[0m\033[1;94m {j:.3f} (Forte)\033[0m\n{"-"*36}')
    elif j > 0.1:
        print(f'\033[1m{i}:\033[0m\033[1;92m {j:.3f} (Médio)\033[0m\n{"-"*36}')
    elif j > 0.02:
        print(f'\033[1m{i}:\033[0m\033[1;93m {j:.3f} (Fraco)\033[0m\n{"-"*36}')
    else:
        print(f'\033[1m{i}:\033[0m\033[1;91m {j:.3f} (Inútil)\033[0m\n{"-"*36}')
````

````python
graf3.sort_values(by='Total', axis=1, ascending=False, inplace=True)
graf3.drop('Total', axis=1).drop('Total').plot.bar(figsize=(10,5), edgecolor='black', stacked=True)
plt.ylabel('Proporção em %')
plt.title('Proporção de notas com crédito por categoria por período', fontsize=14)
plt.legend(bbox_to_anchor=(1, 1))
````

## Avalie o WOE e o IV no tempo

Avalie o WOE e o IV no tempo da variável ```categoria```.

````python
WOE_cat_tempo = []
IV_cat_tempo = []

trimestres = df['Trimestre'].sort_values().unique()

for trimestre in trimestres:
    df_trimestre = df[df['Trimestre'] == trimestre]
    tab = IV(df_trimestre['categoria'], df_trimestre['flag_credito'])
    WOE_cat_tempo.append(tab['woe'])
    IV_cat_tempo.append(tab['iv_parcial'])

WOE_cat_tempo = pd.DataFrame(WOE_cat_tempo, index=trimestres).drop('total', axis=1)
IV_cat_tempo = pd.DataFrame(IV_cat_tempo, index=trimestres).drop('total', axis=1).sum(axis=1).to_frame('IV')
WOE_cat_tempo.loc['total', :] = WOE_cat_tempo.sum()
WOE_cat_tempo.sort_values(by='total', ascending=False, axis=1, inplace=True)
WOE_cat_tempo
````

````python
print(f'\033[1;7mInformation Value (IV):\033[0m\n{"-"*36}')

for i, j in zip(IV_cat_tempo.index, IV_cat_tempo.IV):
    if j > 0.5:
        print(f'\033[1m{i}:\033[0m\033[1;91m {j:.3f} (Suspeito de tão alto)\033[0m\n{"-"*36}')
    elif j > 0.3:
        print(f'\033[1m{i}:\033[0m\033[1;94m {j:.3f} (Forte)\033[0m\n{"-"*36}')
    elif j > 0.1:
        print(f'\033[1m{i}:\033[0m\033[1;92m {j:.3f} (Médio)\033[0m\n{"-"*36}')
    elif j > 0.02:
        print(f'\033[1m{i}:\033[0m\033[1;93m {j:.3f} (Fraco)\033[0m\n{"-"*36}')
    else:
        print(f'\033[1m{i}:\033[0m\033[1;91m {j:.3f} (Inútil)\033[0m\n{"-"*36}')
````

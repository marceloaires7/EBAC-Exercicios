#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# Utilizando a função .rolling() para calcular a Média Movel e o Desvio Padrãõ.round

Ex2 = df['Adj Close'].rolling(90).mean().stack().reset_index()
Ex2.columns = ['Data', 'Ativo', 'MediaMovel']

Ex2['Adj Close'] = (df['Adj Close'].stack()
                    [df['Adj Close'].stack().shape[0]-Ex2.shape[0]:]
                    .reset_index()[0])

Ex2['DesvioPadrao'] = df['Adj Close'].rolling(90).std().stack().reset_index()[0]

Ex2
````

````python
# Criando Gráfico para analisar a MediaMovel ao longo do tempo.

fig = px.line(Ex2, x="Data", y="MediaMovel", color='Ativo', template='plotly_white')
fig.show()
````

````python
# Criando Gráfico para analisar a DesvioPadrao ao longo do tempo.

fig = px.line(Ex2, x="Data", y="DesvioPadrao", color='Ativo', template='plotly_white')
fig.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
# Criação da variável "colunas" para ser utilizado na criação do Gráfico.

lista_indicadores = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']
ativo = ['TSLA']*len(lista_indicadores)

colunas = list(zip(lista_indicadores, ativo))
colunas
````

````python
# Criação da variável "tsla" para separar os dados a serem utilizados no gráfico.

tsla = df.loc['09/20/2021':,colunas]
tsla.columns = lista_indicadores
tasl = tsla.tail(60)
tasl.head()
````

````python
# Dados a serem passados para a função do Gráfico.

graph = {
    'x': tasl.index,
    'open': tasl.Open,
    'close': tasl.Close,
    'high': tasl.High,
    'low': tasl.Low,
    'type': 'candlestick',
    'name': 'TSLA',
    'showlegend': True
}
````

````python
# Criando Gráfico para analisar o CandleStick em um periodo de 60 dias.

layout = Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

grafico = graph_objects.Figure(
    data = [graph],
    layout_title="Ações TSLA (TESLA)",
    layout=layout
)
grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````

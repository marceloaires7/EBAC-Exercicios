```python
# Importação dos Pacotes.

from pandas_datareader import data as web
import yfinance as yfin

import plotly.express as px
from plotly import graph_objects

import pandas as pd
import numpy as np
```

```python
# Coletando Dados das Ações desejadas através do Yahoo Finance.

yfin.pdr_override()
ativos = ['TSLA', 'AAPL', 'AMZN']
dt_ini = '2021-01-01'

yf = web.get_data_yahoo(ativos,
                        dt_ini)

yf.head()
```

```python
fechamento_vertical = yf.stack().reset_index()
fechamento_vertical.columns = ['Data', 'Ativo', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']

fechamento_vertical.head()
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
amplitude_dia = (fechamento_vertical.groupby(['Data', 'Ativo'])[['High', 'Low']]
                                    .mean()
                                    .reset_index()
                                    .assign(Amplitude = lambda x: x['High'] - x['Low']))

amplitude_dia.head()
```

```python
fig = px.line(amplitude_dia, x="Data", y="Amplitude", color='Ativo', template='plotly_white')
fig.show()
```

```python
amplitude_dia.groupby('Ativo')['Amplitude'].mean().sort_values(ascending=False)
```

```python
amplitude_dia.groupby('Ativo')['Amplitude'].sum().sort_values(ascending=False)
```
